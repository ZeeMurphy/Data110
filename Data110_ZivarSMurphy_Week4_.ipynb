```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot") #Sets the style for the plots
```

```python
#Task 1: Basic Bar Chart for Crime Rates in 2023
Offenses_type = ['aggravated Assault', 'Burglary', 'larceny', 'Motor Vehicle Theft', 'Murdr', 'Rape', 'Robbery']
Offenses_per_100k = [264.1, 250.7,1347.2,318.7, 5.7, 38.0, 66.5]
plt.figure(figsize=(12,6))
C = ['orange', 'purple', 'blue', 'green', 'red', 'teal']
bars = plt.bar(Offenses_type, Offenses_per_100k, color=C)
plt.title('Crim rates 2023')
plt.xlabel('Offenses_type', fontsize=12, fontweight='bold')
plt.ylabel('offenses_per_100k', fontsize=12, fontweight='bold')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
```

```python
# Task 2: Change in Crime Rates (2022 vs 2023) with Unique Colors
Change_Offenses_per_100k = [-3.3, -8.1, -4.9, +12.00, -12.0, -9.8, -0.8]
Offenses = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#33FFF1', '#FFBD33', '#A633FF']
plt.figure(figsize=(12,6))
plt.barh(Offenses, Change_Offenses_per_100k, color =colors)
plt.title('Change in Crime Rates (2022 vs. 2023)', fontsize=14, fontweight='bold')
plt.xlabel('percentage_change_in_Offenses', fontsize=12, fontweight='bold')
plt.ylabel('Offense Type', fontsize=12, fontweight='bold')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
```

Why does this bar graph make more sense than Task 1's bar graph?
##**Bar with absolute crime rates for 2023 vs bar with Percentage change over time:**

I) The bar that demonstrates the absolute crime rates for 2023, gives only a snapshot of the total number of crimes pr 100k people. While, the raw numbers don't tell us whether crime rates are increasing or decreasing, the Bar with percentage change, highlights wether offenses have risen or fallen. This provides a better understanding about changes in different types of crimes.
Absolute numbers can be missleading if without taking into consideration the context. For example, in the bar with absolute numbers Larneny has the highest rate and Motor Vehicle Theft is not showing an alarming rate compared to other categories. But the second bar reveals that actually the rate of larsney dropped significantly and the rate of vehicle theft.

II) Policymakers and law enforcement agencies are more interested in trends than in single-year snapshots because trends inform whether their action pland to prevent crimes are working. A rising crime trend (like motor vehicle theft) might prompt stronger measures, while a declining trend (like aggravated assault or burglary) might signal that existing measures are effective.
For the general public, aslo, understanding change over time is more helpful than just seeing the total numbers. If people only see that certain crimes have high rates, they might be alarmed, but understanding that these rates are declining helps paint a clearer, more accurate picture of safety trends in their community.

```python
# Taks 2. Multilple versions

# Data for the chart
Offenses = ['Aggravated Assault', 'Burglary', 'Larceny',  'Motor Vehicle Theft','Murder', 'Rape', 'Robbery']
Change_Offenses_per_100k = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
C = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#33FFF1', '#FFBD33', '#A633FF']

# Create vertical bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(Offenses, Change_Offenses_per_100k, color=C)

# Add titles and labels
plt.title('Change in Crime Rates per 100k People (2022-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Type of Offense', fontsize=12, fontweight='bold')
plt.ylabel('Change in Offenses per 100k', fontsize=12, fontweight='bold')

# Add annotations on the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}',
             va='bottom', ha='center', fontsize=10, color='black')

# Get the current axis using gca()
ax = plt.gca()

# Set y-axis limits to ensure visibility
ax.set_ylim(min(Change_Offenses_per_100k) - 5, max(Change_Offenses_per_100k) + 5)

# Remove specific spines (borders)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add vertical grid lines behind the bars
plt.grid(axis='y', linestyle='--', color='#d3d3d3', zorder=0)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Task 2, Part 2. Improve the clarity, Create multiple versions
# Create a horizontal bar chart and arrange it in acsneding order
data = {'Offenses': ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
        'Change_Offenses_per_100k': [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]}
plt.figure(figsize=(12, 8))
C = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#33FFF1', '#FFBD33', '#A633FF']

dm = pd.DataFrame(data)
dm_sorted = dm.sort_values(by='Change_Offenses_per_100k', ascending=True)
plt.barh(dm_sorted['Offenses'], dm_sorted['Change_Offenses_per_100k'], color=C)

# Add titles and labels
plt.title('Crime Rate/100k people 2022-2023', fontsize=16, fontweight='bold')
plt.xlabel('Change in Offenses per 100k', fontsize=12, fontweight='bold')
plt.ylabel('Offenses', fontsize=12, fontweight='bold')

# Get the current axis using gca()
ax = plt.gca()

# Set x-axis limits to ensure 0 is included (since it's a horizontal bar chart, the x-axis shows the amounts)
ax.set_xlim(min(dm_sorted['Change_Offenses_per_100k']) - 5, max(dm_sorted['Change_Offenses_per_100k']) + 5)

# Remove specific spines (borders)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add horizontal grid lines behind the bars (affects vertical grid for horizontal bars)
plt.grid(axis='x', linestyle='-', color='#d3d3d3', zorder=0)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023
# Sample data
plt.figure(figsize=(10, 5))
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
# Generate indexes for the x-axis (equal to the length of '7' categories)
x_indexes = np.arange(len(x))
print(x_indexes)
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5] #Crime Rate in 2023
z = [273.1, 272.8, 1416.2, 284.6, 6.48, 42.13, 67.04]  # Crime Rate in 2022

# Plot the first set of bars
plt.bar(x_indexes-0.125,y, color='blue', width=0.25, label='Crime Rate 2023')

# Plot the second set of bars
plt.bar(x_indexes + 0.125,z , color='orange', width=0.25, label='Crime Rate 2022')

# Customize the x-axis ticks by setting them to the categories in 'x'
plt.xticks(ticks=x_indexes, labels=x)

plt.xlabel('Offenses Types', fontsize=12, fontweight='bold' )
plt.ylabel('Crime Rates per 100,000 People', fontsize=12, fontweight='bold' )
plt.title('Crime Rates Comparison 2022 vs 2023', color='Red', fontweight='bold')
plt.grid(axis='y', linestyle=('-.'), zorder=0 , linewidth=0.5, alpha=0.25)

plt.legend()

# Display the plot
plt.show()
```
