```python
# 1. Setup and Style
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
additional_style = 'classic'  # Example additional style
```

```python
# 2. Basic Line Plot
#range of x values
x = np.linspace(-2,4)
#define the function
f = x**2 - 2*x + 1
plt.plot(x, f)
```

```python
# 3. Multiple Lines
#define the functions
f = x**3     #f(x)=X^3
g= np.exp(x) #g(x)=e^x
h=np.log(x)  #h(x)=log(x)
plt.plot(x, f, label='f(x) = x³', color='blue', linewidth=2)
plt.plot(x, g, label='g(x) = e^x', color='orange', linewidth=2)
plt.plot(x, h, label='h(x) = log(x)', color='green', linewidth=2)
plt.title('Plot of Three Functions')
plt.xlabel('x')
plt.ylabel('f(x), g(x), h(x)')
plt.legend()
```

```python
# 4. Styling Lines
# Define the functions
f = x**2    # f(x) = x²
g = x**3    # g(x) = x³
h = x**4    # h(x) = x⁴
i = x**5    # i(x) = x⁵
# Create the plot with different color specifications
plt.plot(x, f, color='blue', label='f(x) = x²')            # Named color
plt.plot(x, g, color='#FFA500', label='g(x) = x³')        # Hex color
plt.plot(x, h, color=(0, 1, 0), label='h(x) = x⁴')        # RGB tuple
plt.plot(x, i, color='0.5', label='i(x) = x⁵')             # Grayscale
plt.title('Plot of Polynomial Functions with Different Color Specifications')
plt.xlabel('x')
plt.ylabel('f(x), g(x), h(x), i(x)')
plt.legend()
```
