```python
# 1. Setup and Style
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
# 2. Basic Line Plot
#range of x values
x = np.linspace(-2,4)
#define the function
f = x**2 - 2*x + 1
plt.plot(x, f)
```

```python
# 3. Multiple Lines
#Define the range for x
x = np.linspace(-2, 2, 400)
#define the functions
f_x = x**3     #f(x)=X^3
g_x = np.exp(x) #g(x)=e^x
h_x = np.log(x)  #h(x)=log(x)
plt.plot(x, f_x, label='f(x) = x³', color='blue', linewidth=2)
plt.plot(x, g_x, label='g(x) = e^x', color='orange', linewidth=2)
plt.plot(x, h_x, label='h(x) = log(x)', color='green', linewidth=2)
plt.title('Plot of Three Functions')
plt.xlabel('x')
plt.ylabel('f(x), g(x), h(x)')
plt.legend()
```

```python
# 4. Styling Lines
# Define the range for x for a better viewing
x = np.linspace(-2, 2, 400)
# Define the functions
f_x= x**2    # f(x) = x²
g_x= x**3    # g(x) = x³
h_x= x**4    # h(x) = x⁴
i_x= x**5    # i(x) = x⁵
# the plot with different color specifications and line style
plt.plot(x, f_x, color='blue', label='f(x) = x²', linestyle='-')            # Named color
plt.plot(x, g_x, color='#FFA500', label='g(x) = x³', linestyle='--')        # Hex color
plt.plot(x, h_x, color=(0, 1, 0), label='h(x) = x⁴', linestyle='-.')        # RGB tuple
plt.plot(x, i_x, color='0.5', label='i(x) = x⁵', linestyle=':')             # Grayscale
plt.title('Plot of Polynomial Functions with Different Color Specifications')
plt.xlabel('x')
plt.ylabel('f(x), g(x), h(x), i(x)')
plt.legend()
```

```python
# 5. Axes Limits and Scaling
# The range of x values
x = np.linspace(-10, 10, 400)
f_x = 1/(x**2 + 1) #Function
plt.figure(figsize=(12,5))
# 1
plt.plot(x, f_x, label='f(x) = 1/(x²+1)', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)=1/(x²+1)')
plt.xlim(-10,10)   #Custom x-limits
plt.ylim(0,1.1)    #Custom y-limits
plt.grid(True)
plt.legend()
plt.show()
# the second pllot with reversed x and y axes ('tight')
plt.figure(figsize=(8,6))
plt.plot(f_x, x, label='reversed axes', color='green')
plt.xlabel('f(x)')
plt.ylabel('x')
plt.title('Plot of f(x) = 1/(x²+1) with Reversed Axes')
plt.grid(True)
plt.legend()
plt.axis('tight')
plt.show()
# Another plot demonstrating plt.axis('equal')
plt.figure(figsize=(8,6))
plt.plot(x, f_x, label='f(x)=1/(x²+1)', color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = 1/(x²+1) with Axix equal')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
# 6. Labels and Legends
x = np.linspace(0, 10, 100)
f_x = np.sqrt(x)
g_x = x**1/3
plt.figure(figsize=(10,6))
plt.plot(x, f_x, label=r'f(x) = \sqrt{x}', color='blue')
plt.plot(x, g_x, label=r'g(x) = x^{1/3}$', color='orange')
plt.title("comparaison of $f(x) = \sqrt{x}$ and $g(x)= x^{1/3}$")
plt.legend()
```

```python
# 7. Basic Scatter Plot
# with plt.plot() with 'o' marker
x = np.random.rand(50)  # 50 random x coordinates
y = np.random.rand(50 ) # 50 random y coordinates
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)  #1 row, 2 column, 1st subplot
plt.plot(x,y, 'o')
plt.title('Scatter Plot using plt.plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0,1)
plt.ylim(0,1)
plt.grid()

# with plt.scatter()
plt.subplot(1,2,2)  # 1 row, 2 columns, 2nd supplot
plt.scatter(x,y, color='orange')
plt.title('Scatter Plot using plt.scatter()')
plt.xlabel('X=axis')
plt.ylabel('Y=axis')
plt.xlim(0,1)
plt.ylim(0.1)
plt.grid()
# It seems to me like the visualizations of these two plots are a bit different.
```

```python
# 8 Customizing Scatter Plots
x = np.linspace(0, 4* np.pi, 100) # 100 points from 0, to 4π
f_x = x * np.sin(x)  # f(x) = x * sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, marker='o', markersize=8, linestyle='-', color='blue', linewidth=2, markerfacecolor='orange', markeredgecolor='black', markeredgewidth=1.5)
plt.title('Scatter Plot of f(x) = x * sin(x)', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.grid()
plt.xlim(0, 4 *np.pi)
plt.ylim(min(f_x) - 1, max(f_x) + 1)
```

```python
# 9 Combined Plot
x = np.linspace(-2*np.pi, 2*np.pi, 400)
f_x = x* np.cos(x)
scatter_x = np.arange(-2 * np.pi, 2 * np.pi + np.pi/4, np.pi/4)
scatter_f_x = scatter_x * np.cos(scatter_x)
plt.figure(figsize=(10,6)) #Create the plot
plt.plot(x, f_x, color='blue', linestyle='-', linewidth=2, label=r'$f(x) = x \cos(x)$')
plt.scatter(scatter_x, scatter_f_x, color='red', marker='o', s=100, edgecolor='black', label='Scatter Points')
#Customization:
plt.title('plot of f(x) = x cos (x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='grey', linewidth=0.5, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')
plt.grid()
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(min(f_x)-1, max(f_x)+1)
```

```python
#10. Custom Plot with Alternative Style
# Define the range for x
x_tan = np.linspace(-2 * np.pi, 2 * np.pi, 400)  # For tan(x)
x_ge = np.linspace(-3, 3, 400)  # For g(x) and h(x)

# Define the functions
f_x = np.tan(x_tan)  # f(x) = tan(x)
g_x = x_ge * np.exp(-x_ge**2)  # g(x) = x * e^(-x^2)
h_x = np.sinc(x_ge / np.pi)  # h(x) = sin(x)/x; numpy's sinc function normalizes by pi

# Create the plot with a dark background
plt.style.use('dark_background')

# Create a smaller figure
plt.figure(figsize=(8, 6))  # Adjust the figure size here

# Plotting tan(x)
plt.plot(x_tan, f_x, color='cyan', linestyle='-', linewidth=1.5, label=r'$f(x) = \tan(x)$')

# Plotting g(x) = x * e^(-x^2)
plt.plot(x_ge, g_x, color='magenta', linestyle='--', linewidth=1.5, label=r'$g(x) = x e^{-x^2}$')

# Plotting h(x) = sin(x)/x
plt.plot(x_ge, h_x, color='yellow', linestyle=':', linewidth=1.5, label=r'$h(x) = \frac{\sin(x)}{x}$')
# Add scatter points for g(x) at specific intervals
scatter_x = np.arange(-3, 3.5, 0.5)
scatter_g_x = scatter_x * np.exp(-scatter_x**2)
plt.scatter(scatter_x, scatter_g_x, color='red', marker='o', s=50, edgecolor='black', label='Scatter Points for g(x)')  # Reduce point size

# Customize the plot
plt.title('Combined Functions with Scatter Points', fontsize=16)  # Reduce title size
plt.xlabel('x', fontsize=12)  # Reduce label size
plt.ylabel('Function Value', fontsize=12)  # Reduce label size
plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')  # Add x-axis
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')  # Add y-axis
plt.xlim(-3.5, 3.5)
plt.ylim(-10, 2)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=10)  # Adjust legend size
plt.tight_layout()
```

```python
#11. Style Comparison
# Define the range for x
x = np.linspace(0, 10, 100)
f_x = np.sqrt(x)  # f(x) = sqrt(x)
g_x = x**(1/3)    # g(x) = x^(1/3)

# Function to create the plot in different styles
def create_plot(style):
    plt.style.use(style)  # Set the style
    plt.figure(figsize=(10, 6))

    # Plotting f(x) and g(x)
    plt.plot(x, f_x, label=r'$f(x) = \sqrt{x}$', color='blue')
    plt.plot(x, g_x, label=r'$g(x) = x^{1/3}$', color='orange')

    # Customize the plot
    plt.title("Comparison of $f(x) = \sqrt{x}$ and $g(x) = x^{1/3}$")
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Function Value', fontsize=12)
    plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')  # Add x-axis
    plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')  # Add y-axis
    plt.xlim(0, 10)
    plt.ylim(0, 4)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Show the plot
    plt.show()

# Plot in different styles
create_plot('default')             # Default style
create_plot('seaborn-whitegrid')   # Seaborn white grid style
create_plot('dark_background')      # Alternative dark background style
```

##**Explanation of Styles**
**Default Style:**
Standard white background with basic grid lines. The colors are clear but may not be as vibrant.
This plot style is ideal for academic and professional presentations where clarity is essential.

**Seaborn Whitegrid:**

Enhanced grid lines with a white background. The plot looks cleaner and is visually appealing.
This style is suitable for exploratory data analysis and visualizations where aesthetics matter without compromising clarity.

**Dark Background Style:**
Appearance: A striking contrast with bright colors against a dark background. This style emphasizes the lines and makes them pop.
Use Case: Great for presentations in dark environments or when targeting a more modern, tech-savvy audience.
